# Seaborn Roadmap


```python
#   Types of Functions

#            Figure Level
#            Axis Level
# Main Classification

# 1 Relational Plot
# 2 Distribution Plot
# 3 Categorical Plot
# 4 Regression Plot
# 5 Matrix Plot
# 6 Multiplots
```

```python
# https://seaborn.pydata.org/api.html

```

| Feature                    | Figure-level Functions                             | Axis-level Functions                           |
| -------------------------- | -------------------------------------------------- | ---------------------------------------------- |
| 🔹 Definition              | High-level functions that manage the entire figure | Low-level functions that work on a single axis |
| 🔹 Control                 | Automatically creates figure & axes                | You control axes manually                      |
| 🔹 Flexibility             | Less flexible                                      | More flexible                                  |
| 🔹 Use case                | Quick visualization                                | Custom layouts (subplots, grids)               |
| 🔹 Returns                 | `FacetGrid` / `JointGrid`                          | `Axes` object                                  |
| 🔹 Multi-plots             | Supports multiple plots (facets)                   | Usually one plot per axis                      |
| 🔹 Customization           | Limited                                            | Highly customizable                            |
| 🔹 Requires `plt.figure()` | ❌ No                                               | ✅ Yes (often)                                  |

| Category             | Definition                                         | Purpose                                            | Examples                                              |
| -------------------- | -------------------------------------------------- | -------------------------------------------------- | ----------------------------------------------------- |
| 🔗 Relational Plot   | Shows relationship between two numerical variables | To understand correlation/trends between variables | `scatterplot()`, `lineplot()`, `relplot()`            |
| 📉 Distribution Plot | Shows how data is distributed                      | To analyze spread, frequency, and shape of data    | `histplot()`, `kdeplot()`, `displot()`                |
| 📊 Categorical Plot  | Compares data across categories                    | To compare groups or categories                    | `barplot()`, `boxplot()`, `violinplot()`, `catplot()` |
| 📈 Regression Plot   | Shows relationship with a fitted regression line   | To analyze trends and predictions                  | `regplot()`, `lmplot()`                               |
| 🔲 Matrix Plot       | Visualizes matrix-like data (2D data)              | To show correlations or patterns in matrix form    | `heatmap()`, `clustermap()`                           |
| 🧩 Multiplots        | Creates multiple plots in a grid                   | To compare multiple subsets of data                | `FacetGrid`, `pairplot()`, `PairGrid`                 |


```python
# 🔗 Relational → relation between variables
# 📉 Distribution → data spread
# 📊 Categorical → compare groups
# 📈 Regression → trend + prediction
# 🔲 Matrix → correlation view
# 🧩 Multiplots → many plots together
```

# Distribution Plots (Seaborn)

```python
# Distribution plots like histplot, kdeplot, and rugplot are used for univariate analysis to understand the spread, shape, and density of data
```

```python
# Distribution plots help you understand how your data is spread, shaped, and concentrated
```

| Category    | Concept           | Definition                            | What it tells                 | Plot Used  | Example                |
| ----------- | ----------------- | ------------------------------------- | ----------------------------- | ---------- | ---------------------- |
| 📌 Overview | Distribution Plot | Used for univariate analysis          | Understand how data is spread | —          | —                      |
| 📊 Analysis | Range             | Difference between min and max values | Spread of data                | Histplot   | `sns.histplot(x=data)` |
| 📍 Analysis | Central Tendency  | Mean, Median, Mode                    | Center of data                | KDE / Hist | `sns.kdeplot(x=data)`  |
| 📈 Analysis | Shape             | Normal, skewed, bimodal               | Pattern of distribution       | KDE        | `sns.kdeplot(x=data)`  |
| 🔄 Analysis | Spread            | Variance / Std deviation              | Data variability              | Histplot   | `sns.histplot(x=data)` |
| ⚠️ Analysis | Outliers          | Extreme values                        | Unusual data points           | Rugplot    | `sns.rugplot(x=data)`  |


```python
# Plot types
```

| Feature          | How                 |
| ---------------- | ------------------- |
| 📏 Range         | Width of histogram  |
| 📍 Central value | Peak of KDE         |
| 📈 Skewness      | Left/right tilt     |
| 🔁 Bimodal       | Two peaks in KDE    |
| ⚠️ Outliers      | Isolated rug points |

```python
# hue -> fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=10,aspect=2)
plt.show()
```

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
plt.show()
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count of observations within each rectangle with the fill color

# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
plt.show()
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip')
plt.show()
```

| Plot             | Definition                                                      | Key Parameters     | What it Shows                           | When to Use                                                | Example                                                        |
| ---------------- | --------------------------------------------------------------- | ------------------ | --------------------------------------- | ---------------------------------------------------------- | -------------------------------------------------------------- |
| 📊 **displot()** | Figure-level function for distribution (histogram/KDE combined) | `hue`, `kde=True`  | Full distribution + grouping            | When you want **automatic plots + multiple distributions** | `sns.displot(data=tips, x='total_bill', hue='sex', kde=True)`  |
| 📈 **kdeplot()** | Smooth density curve of data                                    | `hue`, `fill=True` | Shape of distribution (peaks, skewness) | When you want **smooth visualization**                     | `sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True)` |
| 📍 **rugplot()** | Shows individual data points as small ticks                     | `hue` (optional)   | Exact positions of data                 | When you want to see **data density + outliers**           | `sns.rugplot(data=tips, x='total_bill')`                       |


| Plot    | Style           | Best For             |
| ------- | --------------- | -------------------- |
| displot | Histogram + KDE | Overall distribution |
| kdeplot | Smooth curve    | Shape & peaks        |
| rugplot | Small ticks     | Exact values         |


```python
# displot = full distribution, kdeplot = smooth shape, rugplot = exact data points
```

# Matrix Plot

```python
# What is a Matrix Plot?

#  A matrix plot is used to visualize 2D data (rows × columns)
#  Mostly used for:

# Correlation matrices
# Pattern detection
# Feature relationships
```

| Plot          | Definition           | Purpose                    | Data Type        | Function           |
| ------------- | -------------------- | -------------------------- | ---------------- | ------------------ |
| 🔥 Heatmap    | Color-coded matrix   | Show values using colors   | 2D data (matrix) | `sns.heatmap()`    |
| 🧩 Clustermap | Heatmap + clustering | Group similar rows/columns | 2D data          | `sns.clustermap()` |


```python
# Heatmap

# Plot rectangular data as a color-encoded matrix
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')

# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# annot
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5, cmap='summer')
plt.show()
```

```python
# Clustermap = Heatmap + Hierarchical Clustering
```
