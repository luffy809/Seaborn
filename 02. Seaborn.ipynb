```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add species column
df['species'] = iris.target

# Map numeric → names
dict_data = {
    0: 'red',      # setosa
    1: 'green',    # versicolor
    2: 'blue'      # virginica
}

# Select features
x = df['petal length (cm)']
y = df['petal width (cm)']

# Scatter plot
plt.scatter(x, y, c=df['species'].apply(lambda x: dict_data[x]))

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")

plt.show()
```

# sns.barplot()

```python
# sns.barplot(x=None, y=None, data=None, hue=None, estimator='mean')
```

```python
# x → categories (e.g., day, product, month)
# y → numeric values (e.g., sales, marks)
# data → dataset (DataFrame)
# hue → grouping (extra category → multiple bars)
# estimator → what to calculate (default = mean)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.barplot(
    data=tips,
    x="day",
    y="total_bill"
)

plt.title("Average Bill per Day")
plt.show()
```

```python
sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)

plt.title("Average Bill by Day & Gender")
plt.show()
```

```python
# x="day"
# Groups data by days
# So you get: Thur, Fri, Sat, Sun
```

```python
# y="total_bill"
# Takes all bill values for each day
```

```python
# barplot logic (VERY IMPORTANT)

#  For each group:

# It calculates the average (mean) of total_bill

# So:

# For Thursday → average bill
# For Friday → average bill
# etc.
```

```python
# # Let’s take Friday:

# Blue bar (Male) ≈ higher
# Orange bar (Female) ≈ lower

#  Meaning:

# On Friday, men spent more on average than women
```

| Day | Sex    | Bill |
| --- | ------ | ---- |
| Fri | Male   | 20   |
| Fri | Male   | 18   |
| Fri | Female | 14   |
| Fri | Female | 16   |


```python
import numpy as np

sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    estimator=np.sum
)

plt.title("Total Bill per Day (Sum)")
plt.show()
```

```python
# Now bars show sum instead of average
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

ax = sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)

# Add numbers on bars
for container in ax.containers:
    ax.bar_label(container)

plt.title("Average Bill by Day & Gender")
plt.show()
```

```python
# ax.containers → gets all bars
#  ax.bar_label() → puts numbers on top of bars
```

```python
for container in ax.containers:
    ax.bar_label(container, fmt="%.1f")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

ax = sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)

for container in ax.containers:
    ax.bar_label(container, fmt="%.1f")

plt.title("Average Bill by Day & Gender")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create dataset
data = {
    "Student": ["A","B","C","A","B","C"],
    "Subject": ["Math","Math","Math","Science","Science","Science"],
    "Marks": [85,78,92,88,76,95]
}

df = pd.DataFrame(data)

# Plot
ax = sns.barplot(
    data=df,
    x="Subject",
    y="Marks"
)

# Add labels
for container in ax.containers:
    ax.bar_label(container, fmt="%.1f")

plt.title("Average Marks by Subject")
plt.show()
```

| Subject | Calculation        | Average |
| ------- | ------------------ | ------- |
| Math    | (85 + 78 + 92) / 3 | 85.0    |
| Science | (88 + 76 + 95) / 3 | 86.3    |
